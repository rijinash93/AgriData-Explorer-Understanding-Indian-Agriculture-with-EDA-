```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data.csv')
```

```python
df
```

```python
df.isnull().sum
```

```python
df.dropna()
df.dropna(axis=1)
```

```python
df= df.fillna("")
```

```python
df
```

```python
# Sort by rice production and get top 7
top7_rice = df.sort_values(by='RICE PRODUCTION (1000 tons)', ascending=False).head(7)
top7_rice
```

```python
print(df.columns.tolist())
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top7_rice['Year'], top7_rice['RICE PRODUCTION (1000 tons)'], color='green')
plt.title('Top 7 Rice Producing States')
plt.xlabel('Year')
plt.ylabel('Rice Production (in tonnes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top5_wheat = df.groupby('State Name')['WHEAT PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(5)

# Bar Chart
top5_wheat.plot(kind='bar', color='orange', figsize=(8,5))
plt.title('Top 5 Wheat Producing States')
plt.ylabel('Wheat Production (in tonnes)')
plt.xticks(rotation=45)
plt.show()

# Pie Chart
top5_wheat.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title('Wheat Production % by Top 5 States')
plt.ylabel('')
plt.show()
```

```python
top5_oil = df.groupby('State Name')['OILSEEDS PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(5)
top5_oil.plot(kind='bar', color='brown', figsize=(8,5))
plt.title('Top 5 Oil Seed Producing States')
plt.ylabel('Oilseed Production (in tonnes)')
plt.xticks(rotation=45)
plt.show()
```

```python
top7_sunflower = df.groupby('State Name')['SUNFLOWER PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(7)
top7_sunflower.plot(kind='bar', color='yellow', figsize=(8,5))
plt.title('Top 7 Sunflower Producing States')
plt.ylabel('Sunflower Production (in tonnes)')
plt.xticks(rotation=45)
plt.show()
```

```python
sugarcane_50yr = df.groupby('Year')['SUGARCANE PRODUCTION (1000 tons)'].sum().tail(50)
sugarcane_50yr.plot(kind='line', marker='o', color='purple', figsize=(10,6))
plt.title("India's Sugarcane Production Over Last 50 Years")
plt.ylabel('Sugarcane Production (in tonnes)')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

```python
last50 = df.groupby('Year')[['RICE PRODUCTION (1000 tons)', 'WHEAT PRODUCTION (1000 tons)']].sum().tail(50)
last50.plot(kind='line', figsize=(10,6), marker='o')
plt.title('Rice vs Wheat Production (Last 50 Years)')
plt.ylabel('Production (in tonnes)')
plt.xlabel('Year')
plt.grid(True)
plt.legend(['Rice', 'Wheat'])
plt.show()
```

```python
wb_rice = df[df['State Name'] == 'West Bengal'].groupby('Dist Name')['RICE PRODUCTION (1000 tons)'].sum().sort_values(ascending=False)
wb_rice.plot(kind='bar', color='teal', figsize=(10,6))
plt.title('Rice Production by Districts in West Bengal')
plt.ylabel('Rice Production (in tonnes)')
plt.xticks(rotation=90)
plt.show()
```

```python
up_wheat = df[df['State Name'] == 'Uttar Pradesh'].groupby('Year')['WHEAT PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(10)
up_wheat.plot(kind='bar', color='blue', figsize=(10,5))
plt.title('Top 10 Years of Wheat Production from Uttar Pradesh')
plt.ylabel('Wheat Production (in tonnes)')
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.show()
```

```python
df['MILLET PRODUCTION'] = (
    df['PEARL MILLET PRODUCTION (1000 tons)'] + df['FINGER MILLET PRODUCTION (1000 tons)']
) * 1000

millet_50 = df.groupby('Year')['MILLET PRODUCTION'].sum().tail(50)
millet_50.plot(kind='line', marker='o', color='green', figsize=(10,6))
plt.title("Millet Production in India (Last 50 Years)")
plt.ylabel("Production (Tonnes)")
plt.xlabel("Year")
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
sorghum_kharif = df.groupby('State Name')['KHARIF SORGHUM PRODUCTION (1000 tons)'].sum()
sorghum_rabi = df.groupby('State Name')['RABI SORGHUM PRODUCTION (1000 tons)'].sum()

sorghum_by_state = pd.DataFrame({
    'KHARIF': sorghum_kharif,
    'RABI': sorghum_rabi
}) * 1000

sorghum_by_state = sorghum_by_state.sort_values(by='KHARIF', ascending=False).head(10)
sorghum_by_state.plot(kind='bar', stacked=True, figsize=(12,6), colormap='Set2')
plt.title('Sorghum Production (Kharif vs Rabi) by Top 10 States')
plt.ylabel('Production (Tonnes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
groundnut = df.groupby('State Name')['GROUNDNUT PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(7) * 1000
groundnut.plot(kind='bar', color='saddlebrown', figsize=(10,6))
plt.title("Top 7 States for Groundnut Production")
plt.ylabel("Production (Tonnes)")
plt.xlabel("State")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
soybean = df.groupby('State Name')[['SOYABEAN PRODUCTION (1000 tons)', 'SOYABEAN AREA (1000 ha)']].sum()
soybean['YIELD'] = (soybean['SOYABEAN PRODUCTION (1000 tons)'] * 1000) / (soybean['SOYABEAN AREA (1000 ha)'] * 1000)
top5_soybean = soybean.sort_values(by='SOYABEAN PRODUCTION (1000 tons)', ascending=False).head(5)

# Production Bar Chart
top5_soybean['SOYABEAN PRODUCTION (1000 tons)'].plot(kind='bar', color='orange', figsize=(8,5))
plt.title("Top 5 Soybean Producing States")
plt.ylabel("Production (1000 Tons)")
plt.tight_layout()
plt.show()

# Yield Efficiency
top5_soybean['YIELD'].plot(kind='bar', color='olive', figsize=(8,5))
plt.title("Soybean Yield Efficiency (Tonnes per Hectare)")
plt.ylabel("Yield")
plt.tight_layout()
plt.show()
```

```python
oilseed = df.groupby('State Name')['OILSEEDS PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(10) * 1000
oilseed.plot(kind='bar', color='goldenrod', figsize=(10,6))
plt.title("Top 10 Oilseed Producing States")
plt.ylabel("Production (Tonnes)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
subset = df[['RICE AREA (1000 ha)', 'RICE PRODUCTION (1000 tons)',
             'WHEAT AREA (1000 ha)', 'WHEAT PRODUCTION (1000 tons)',
             'MAIZE AREA (1000 ha)', 'MAIZE PRODUCTION (1000 tons)']]

melted = pd.DataFrame({
    'AREA': pd.concat([df['RICE AREA (1000 ha)'], df['WHEAT AREA (1000 ha)'], df['MAIZE AREA (1000 ha)']]) * 1000,
    'PRODUCTION': pd.concat([df['RICE PRODUCTION (1000 tons)'], df['WHEAT PRODUCTION (1000 tons)'], df['MAIZE PRODUCTION (1000 tons)']]) * 1000,
    'CROP': ['Rice'] * len(df) + ['Wheat'] * len(df) + ['Maize'] * len(df)
})

sns.lmplot(data=melted, x='AREA', y='PRODUCTION', hue='CROP', height=6, aspect=1.5)
plt.title("Impact of Area on Production (Rice, Wheat, Maize)")
plt.xlabel("Area (Hectares)")
plt.ylabel("Production (Tonnes)")
plt.tight_layout()
plt.show()
```

```python
yield_df = df.groupby('State Name')[['RICE YIELD (Kg per ha)', 'WHEAT YIELD (Kg per ha)']].mean().dropna()
yield_df.plot(kind='bar', figsize=(12,6))
plt.title("Average Rice vs Wheat Yield Across States")
plt.ylabel("Yield (Kg per Hectare)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
